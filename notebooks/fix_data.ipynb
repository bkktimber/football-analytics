```python
lst = []
for e in pd.read_pickle('/Users/mai/Projects/football-analytics/data/whoscored/epl/20202021/1485563.pkl')['events']:
    for item in e['qualifiers']:
        if item['type']['displayName'] == 'Angle':
            m = e['minute']
            s = e['second']
            x1 = e['x']
            y1 = e['y']
            x2 = e['endX']
            y2 = e['endY']
            a = item['value']
            lst.append([m,s,x1,y1,x2,y2,float(a)])
tmp = pd.DataFrame(lst, columns=['min', 'sec', 'x1', 'y1',  'x2', 'y2', 'angle'])
tmp.head()
```
